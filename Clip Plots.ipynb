```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
giaSitting = pd.read_csv("Chris Running.csv", header = None, skiprows = 2)
giaStanding = pd.read_csv("Chris Walking.csv", header = None, skiprows = 2)
```

```python
def createDF(sub, act, data):
    data = data.drop(4,1)
    x = 0
    i = 0
    array = []
    d = []
    for index, row in data.iterrows():
        if (data.iloc[index,0] %5 != 0):
            array.append([data.iloc[index,1],data.iloc[index,2],data.iloc[index,3]])
        elif (data.iloc[index,0] %5 == 0):
            d.append([sub , act, array])
            array = []
            i+=1

    d.pop(0)
    d.pop(1)
    d.pop(2)
    d.pop(3)
    d.pop(-1)
    d.pop(-2)

    dftest = pd.DataFrame(d)
    
    return dftest
```

```python
df_sit = createDF("gia", "walking", giaSitting)
df_stand = createDF("gia", "standing", giaStanding)
```

```python
sitting_clip = df_sit.iloc[0,2]
standing_clip = df_stand.iloc[0,2]
```

```python
t_sit = range(len(sitting_clip))
plt.hold('on')
plt.plot(t_sit, [i[0] for i in sitting_clip], t_sit, [i[1] for i in sitting_clip], t_sit, [i[2] for i in sitting_clip])
plt.legend(['x', 'y', 'z'])
plt.title('walking')
plt.hold('off')
```

```python

t_stand = range(len(standing_clip))
plt.hold('on')
plt.plot(t_stand, [i[0] for i in standing_clip], t_stand, [i[1] for i in standing_clip], t_stand, [i[2] for i in standing_clip])
plt.legend(['x', 'y', 'z'])
plt.title('standing')
plt.hold('off')
```
