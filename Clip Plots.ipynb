```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
giaRunning = pd.read_csv("data/Gia Running.csv", header = None, skiprows = 2)
giaWalking = pd.read_csv("data/Gia Walking.csv", header = None, skiprows = 2)
giaStanding = pd.read_csv("data/Gia Standing.csv", header = None, skiprows = 2)
giaSitting = pd.read_csv("data/Gia Sitting.csv", header = None, skiprows = 2)
giaClimbing = pd.read_csv("data/Gia Climbing.csv", header = None, skiprows = 2)
giaSleeping = pd.read_csv("data/Gia Sleeping.csv", header = None, skiprows = 2)
```

```python
def createDF(sub, act, data):
    data = data.drop(4,1)
    x = 0
    i = 0
    array = []
    d = []
    for index, row in data.iterrows():
        if (data.iloc[index,0] %5 != 0):
            array.append([data.iloc[index,1],data.iloc[index,2],data.iloc[index,3]])
        elif (data.iloc[index,0] %5 == 0):
            d.append([sub , act, array])
            array = []
            i+=1

#     d.pop(0)
#     d.pop(-1)

    dftest = pd.DataFrame(d)
    
    return dftest
```

```python
df_sit = createDF("gia", "sitting", giaSitting)
df_stand = createDF("gia", "standing", giaStanding)
df_walk = createDF("gia", "walking", giaWalking)
df_run = createDF("gia", "running", giaRunning)
df_climb = createDF("gia", "climbing", giaClimbing)
df_sleep = createDF("gia", "sleeping", giaSleeping)
```

```python
sitting_clip = df_sit.iloc[2,2]
standing_clip = df_stand.iloc[2,2]
walking_clip = df_walk.iloc[2,2]
running_clip = df_run.iloc[2,2]
climbing_clip = df_climb.iloc[1,2]
sleeping_clip = df_sleep.iloc[2,2]
```

```python
def plot_clip (clip, title):
    t = range(len(clip))
    plt.hold('on')
    plt.plot(t, [i[0] for i in clip], t, [i[1] for i in clip], t, [i[2] for i in clip])
    plt.legend(['x', 'y', 'z'])
    plt.xlabel('time, ms')
    plt.ylabel('acceleration, m/s2')
    plt.title(title)
    plt.hold('off')
```

```python
plot_clip(sitting_clip, 'Sitting')
```

```python
plot_clip(standing_clip, 'Standing')
```

```python
plot_clip(walking_clip, 'Walking')
```

```python
plot_clip(running_clip, 'Running')
```

```python
plot_clip(climbing_clip, 'Climbing')
```

```python
plot_clip(sleeping_clip, 'Sleeping')
```

# Plotting FFT signals

```python
# sample FFt for sitting activity X axis
t = np.arange(len(sitting_clip))
fft = np.fft.fft([i[0] for i in sitting_clip])
freq = np.fft.fftfreq(t.shape[-1])
plt.plot(freq, abs(fft))
plt.show()
```

```python
def plot_fft(clip, act):
    N = len(clip)
    t = np.arange(N)
    t = np.fft.fftfreq(t.shape[-1])
    print(t)
    
    plt.rcParams["figure.figsize"] = (16,5)
    
    plt.subplot(1, 3, 1)
    plt.plot(t, abs(np.fft.fft([i[0] for i in clip])))
    plt.title('FFT X ' + act)
    
    plt.subplot(1, 3, 2)
    plt.plot(t, abs(np.fft.fft([i[1] for i in clip])))
    plt.title('FFT Y ' + act)
    
    plt.subplot(1, 3, 3)
    plt.plot(t, abs(np.fft.fft([i[2] for i in clip])))
    plt.title('FFT Z ' + act)
    
    plt.show()
```

```python
plot_fft(sitting_clip, 'Sitting')
```

```python
plot_fft(standing_clip, 'Standing')
```

```python
plot_fft(walking_clip, 'Walking')
```

```python
plot_fft(running_clip, 'Running')
```

```python
plot_fft(climbing_clip, 'Climbing')
```

```python
plot_fft(sleeping_clip, 'Sleeping')
```
